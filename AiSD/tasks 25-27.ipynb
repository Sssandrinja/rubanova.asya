```python
#25.7
#Класс – квадратное уравнение. Методы – расчет корней уравнения. Поля – корни уравнения, коэффициенты  a, b, c.

class Solve:
    def __init__(self, a, b, c):
        """Инициализация полей класса"""
        self.a = a
        self.b = b
        self.c = c
        self.root1 = None
        self.root2 = None

    def calc(self):
        """Функция, вычисляющая корни квадратного уравнения"""
        discriminant = self.b**2 - 4*self.a*self.c
        if discriminant > 0:
            self.root1 = (-self.b + discriminant**0.5) / (2*self.a)
            self.root2 = (-self.b - discriminant**0.5) / (2*self.a)
        elif discriminant == 0:
            self.root1 = -self.b / (2*self.a)
        else:
            self.root1 = "Не существует"
            self.root2 = "Не существует"

# Пример использования Solve
ex = Solve(1, -3, 2)
ex.calc()
print("Корень 1: ", ex.root1)
print("Корень 2: ", ex.root2)
```

```python
# 26.7
# Составьте описание класса многочлена вида ax2+bx+c. Включите в класс метод вывода описания многочлена на экран и 
# метод вычисления значения многочлена для заданного аргумента, а также свойства, позволяющие: определить, 
# имеет ли квадратное уравнение действительные корни; умножить многочлен на скаляр.

class Polynom:
    def __init__(self, a, b, c):
        # Инициализация полей класса - коэффициентов многочленa
        self.a = a
        self.b = b
        self.c = c

    def __str__(self):
        # Магический метод для представления обьекта класса Polynom в красивом виде при вызове функцией принт например
        return f"{self.a}x^2 + {self.b}x + {self.c}"

    def calc(self, x):
        # Вычисление многочлена при конкретном х
        return self.a*x**2 + self.b*x + self.c

    def has_real_roots(self):
	# Проверка на наличие действительных корней(D>0)
        discriminant = self.b**2 - 4*self.a*self.c
        return discriminant >= 0

    def umn_scalar(self, scalar):
	# Умножение многочлена на скаляр
        self.a *= scalar
        self.b *= scalar
        self.c *= scalar

# Примеры использования каждого из методов
ex = Polynom(1, -3, 2)
print("Описание многочлена: ", ex)
print("Значение многочлена при x = 2: ", ex.calc(2))
print("Имеет ли многочлен действительные корни: ", ex.has_real_roots())
ex.umn_scalar(2)
# описание умноженного многочлена
print(ex)
```

```python
#27.7
#Создайте класс КЛИЕНТ с методами, позволяющими вывести на экран информацию о клиентах банка, 
#а также определить соответствие клиента критерию поиска. Создайте дочерние классы ВКЛАДЧИК 
#(фамилия, дата открытия вклада, размер вклада, процент по вкладу), КРЕДИТОР 
#(фамилия, дата выдачи кредита, размер кредита, процент по кредиту, остаток долга), 
#ОРГАНИЗАЦИЯ (название, дата открытия счета, номер счета, сумма на счету) со своими методами 
#вывода информации на экран и определения соответствия дате (открытия вклада, выдаче кредита, открытия счета). 
#Создайте список из n клиентов, выведите полную информацию из базы на экран, а также организуйте поиск клиентов, начавших сотрудничать с банком в заданную дату

class Client:
    """Родительский(начальный) класс"""
    def __init__(self, last_name):
        self.last_name = last_name

    def display_info(self):
        pass

    def check_criteria(self, search_date):
        pass

class Depositor(Client):
    """Класс вкладчик"""
    def __init__(self, last_name, deposit_open_date, deposit_amount, deposit_interest):
        # Инициализирует одно из полей класса из родительского класса
        super().__init__(last_name)
        self.deposit_open_date = deposit_open_date
        self.deposit_amount = deposit_amount
        self.deposit_interest = deposit_interest

    def display_info(self):
        print(f"Вкладчик: {self.last_name}, Дата открытия вклада: {self.deposit_open_date}, Сумма вклада: {self.deposit_amount}, Процент по вкладу: {self.deposit_interest}%")

    def check_criteria(self, search_date):
        return self.deposit_open_date == search_date

class Creditor(Client):
    def __init__(self, last_name, credit_issue_date, credit_amount, credit_interest, remaining_debt):
        super().__init__(last_name)
        self.credit_issue_date = credit_issue_date
        self.credit_amount = credit_amount
        self.credit_interest = credit_interest
        self.remaining_debt = remaining_debt

    def display_info(self):
        print(f"Кредитор: {self.last_name}, Дата выдачи кредита: {self.credit_issue_date}, Размер кредита: {self.credit_amount}, Процент по кредиту: {self.credit_interest}%, Остаток долга: {self.remaining_debt}")

    def check_criteria(self, search_date):
        return self.credit_issue_date == search_date

class Organization(Client):
    def __init__(self, name, account_open_date, account_number, account_balance):
        super().__init__(name)
        self.account_open_date = account_open_date
        self.account_number = account_number
        self.account_balance = account_balance

    def display_info(self):
        print(f"Организация: {self.last_name}, Дата открытия счета: {self.account_open_date}, Номер счета: {self.account_number}, Сумма на балансе: {self.account_balance}")

    def check_criteria(self, search_date):
        return self.account_open_date == search_date

# Список клиентов
clients = [
    Depositor("Катя", "11.03.2004", 5000, 3.5),
    Creditor("Саша", "24.01.2005", 10000, 5, 7500),
    Organization("ФУ при правительстве РФ", "20.02.2002", "123456", 100000)
]

# Вывод полной информации о клиентах
for client in clients:
    client.display_info()

# Поиск клиентов, начавших сотрудничать с банком в заданную дату
search_date = "11.03.2004"
print("\nКлиенты, начавшие сотрудничать с банком в дату", search_date)
for client in clients:
    if client.check_criteria(search_date):
        client.display_info()
```
